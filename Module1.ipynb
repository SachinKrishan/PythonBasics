```python
def interrupt():
    """
        Prints the following pattern forever (till interrupt):

        *
        00
        ***
        0000
        *****
        0000
        ***
        00
        *

        Ideas
        -----
        Looping, Conditions
    """
    # toolbar: kernel->interrupt to stop
    while True:
        for i in range (9):
            if i<5:
                for j in range (i+1):
                    if (i+1)%2==1:
                        print "*",
                    else:
                        print "0",
            else:
                for j in range (9-i):
                    if (i+1)%2==1:
                        print "*",
                    else:
                        print "0",
            print "\n"
            time.sleep(0.5)
            
            
    
                        
```

```python
# Call your function here

interrupt()

```
