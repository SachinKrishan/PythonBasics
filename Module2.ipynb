```python
# Your function goes here

def prettydf(filename):
    dbfile = open(filename, 'rb')      
    temp   = pkl.load(dbfile)

    temp   = temp['signal']
    ACC    = temp['ACC']
    ECG    = temp['ECG']

    accdf     = pd.DataFrame(ACC,columns = ["ACCx","ACCy","ACCz"])
    newdf    = pd.DataFrame()

    newdf['ACCx']= accdf["ACCx"] 
    newdf['ACCy']= accdf["ACCy"] 
    newdf['ACCz']= accdf["ACCz"]  
    
    newdf['ECG']   = reduce(lambda z, y :z + y, temp['ECG']) 
    newdf['RESP']  = reduce(lambda z, y :z + y, temp['RESP'])
    newdf['LABEL'] = reduce(lambda z, y :z + y, temp['LABEL']) 
   
    newdf.to_csv('pretty.csv', index=False, encoding='utf-8')
    dbfile.close()
    
    return newdf
```

## Task 2

Some simple file operations.

```python
"""
    Generate a text file containing 50 integer elements between 1 and 20 (use `random`). 
    Identify the unique numbers and the number of their occurrences, and write this into 
    a new text file.
    
    Example
    -------
    
    Generated File
    --------------
    1 2 3 4 3 2 1 ...
    
    Count File
    ----------
    1 - 2
    2 - 2
    3 - 2
    4 - 1
    ...
"""
```

```python
# Your function goes here
import random

def CountFrequency(my_list): 
  
    freq = {} 
    for item in my_list: 
        if (item in freq): 
            freq[item] += 1
        else: 
            freq[item] = 1
    return freq
    

def rnd50():
    file1 = open("myfile.txt","w") 

    filelist = [str(random.randrange(1,20))+'\n' for x in range(50)] 
    
    file1.writelines(filelist)

    file1.close()    


    
def rnd50frq():
    file2 = open("myfile.txt","r")

    mylist = file2.readlines()

    final=[]

    for i in range(50):
        final.append(mylist[i].rstrip())

    file2.close()
    
    freq = CountFrequency(final)
    
    
        
    file3 = open("freq.txt","w") 
    for key in freq:
        line = key + "-" + str(freq[key]) + "\n"
        
        file3.write(line)

    file3.close() 
    
```
